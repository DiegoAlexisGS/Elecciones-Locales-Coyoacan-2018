## Seleccionamos solo Coyoacán

```python
dt9.columns
```

```python
Co=dt9[dt9["Delegación "]=="Coyoacán"]
```

```python
len(Co["Sección"])
```

```python
Co
```

```python
partidos=["PAN","PRI","PRD","PT","PVEM","Convergencia","Nueva Alianza","PSD","Votos Nulos","Candidatura Común PRD_PT_Convergencia"]
```

```python
provi=[i for i in list(Co.columns) if type(list(Co[i])[0])!=str]
for i in provi:
    if sum(Co[i])==0:
        Co.drop(i,axis=1,inplace=True)
```

```python
Co
```

```python
for i in partidos:
    Co["%"+i]=(Co[i]/Co["Votación Total"])*100
```

```python
writer=ExcelWriter("/home/diego/Escritorio/Thesis/clean_data/data/elecciones_09_12_15_18/co_2009.xlsx")
Co.to_excel(writer,"hoja de calculo",index=False)
writer.save()
```

## Generar los porcentajes

```python
Partidos=["PAN","PRI","PRD","PT","PVEM","MC","NA","PRI-PVEM CC1","PRD-PT-MC CC2",'VOTOS\nNULOS']
for i in Partidos:
    C12["%"+i]=(C12[i]/C12["VOTACION\nTOTAL"])*100
```
