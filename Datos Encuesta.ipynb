```python
ps=list(zip(secc,par))
```

```python
party=[]
loc=[]
for i in range(len(ps)):
    if len(ps[i][1])>1:
        for j in range(len(ps[i][1].split(','))):
            loc.append(ps[i][0])
            party.append(int(j))
    else:
        loc.append(ps[i][0])
        party.append(int(ps[i][1]))
            
        
```

```python
info=pd.DataFrame(list(zip(loc,party)))
```
